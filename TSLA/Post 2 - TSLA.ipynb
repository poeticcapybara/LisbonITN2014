Let's first import all the necessary modules

```python
import pandas as pd
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt

import pandas.io.data as web
import datetime
```

We define now the start date, end date, ticker(s) and source for the data. In this case we choose roughly 5 years of daily data for the Tesla stock, whose ticker is TSLA. We will use [Yahoo Finance](http://finance.yahoo.com) as the source in the DataReader module 

```python
start_date = datetime.datetime(2009,1,1)
end_date = datetime.datetime.today()
ticker = 'TSLA'
src = 'yahoo'
```

We now download the data and plot the daily close price

```python
underlying = web.DataReader(ticker,src,start_date,end_date)
underlying.Close.plot()
```

### Import from previously downloaded data

```python
fnames = [username+'_tweets.csv' for username in usernames]
dfs = []
for fname in fnames:
    df = pd.read_csv(fname,header=0)
    df['username'] = fname.split('_')[0]
    dfs.append(df)
twdf = pd.concat(dfs)
twdf.set_index('created_at',inplace=True)
twdf.index = pd.to_datetime(twdf.index)
```

```python
data = underlying.join(twdf,how='outer')
del data['id']
```

```python
data_resmpl = data.resample('D',how='mean')
data_resmpl['total_tweets'] = data['text'].groupby(lambda x: x.date).count()
date = data.index.map(lambda x: x.date)
tweets = data.groupby([date,'username']).text.count().unstack()
tweets.index = pd.to_datetime(tweets.index)
data_resmpl = data_resmpl.join(tweets,how='outer')
```

```python
data_resmpl.dropna(how='all',inplace=True)
data_resmpl.ix[:,['Close','total_tweets','SpaceX','elonmusk','TeslaMotors']].plot()
data_resmpl.Close.pct_change().plot(secondary_y=True,label='Return')
plt.legend()
```

```python
s_tweets = data['text']
data['bin_tweet']= s_tweets.apply(lambda x: 0 if str(x)=='nan' else 1)
```

```python
rets = data_resmpl[['Close','Open','High','Low']].pct_change()
rets = rets.join([tweets, data_resmpl.total_tweets], how='outer')
```

```python
%matplotlib
pd.scatter_matrix(rets,diagonal='kde',alpha=0.3)
```

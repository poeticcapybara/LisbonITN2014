```python
import pandas as pd
from lxml import objectify
from collections import defaultdict

import xml.etree.ElementTree as ET

from xml.dom import minidom
```

```python
# For the first time
tree = ET.parse('ptwiktionary-20141020-pages-articles.xml')
root = tree.getroot()
for child in root.iter():
    child.tag = child.tag[42:]
iterroot = root.findall('page')
for result in root.findall('page'):
    ns = result.find('ns')
    if ns.text != '0':
        root.remove(result)
tree.write('output.xml')
```
